Using the equation (2) in page 13 of the following paper:

https://arxiv.org/pdf/1902.10045.pdf


Function for the LogParabola:
(Beta disappears for PowerLaw)

$ \frac{dN}{dE} = K \left ( \frac{E}{E_0} \right )^{-\alpha -\beta\  log (E/ E_0)} $


Differential flux

$ E^2\ \frac{dN}{dE} = E^2\ K \left ( \frac{E}{E_0} \right )^{-\alpha -\beta\  log (E/ E_0)} $

```python
# Functions


# PARAMETERS MUST BE GIVEN IN CGS UNITS!



## dN/dE in 'erg-1 cm-2 s-1'
def dNdE(E, E_0, K, alpha, beta=None):
    
    if (beta == None):
        Y = K * ((E/E_0)**(-alpha))
    else:
        Y = K * ((E/E_0)**(-alpha - beta * np.log(E/E_0)))
        
    return Y



## E^2 dN/dE in 'erg cm-2 s-1'
def diff_flux(E, E_0, K, alpha, beta=None):

    if (beta == None):
        Y = E**2 * K * ((E/E_0)**(-alpha))
    else:
        Y = E**2 * K * ((E/E_0)**(-alpha - beta * np.log(E/E_0)))
        
    return Y
```

### Integrated Flux

```python
# Test by Tarek:
import scipy.integrate as integrate

E0_LP    = (466.0457 * u.MeV).to('erg').value                                           # Pivot_Energy in erg
K_LP     = (3.298064e-10 * u.MeV**-1 * u.cm**-2 * u.s**-1).to('erg-1 cm-2 s-1').value   # LP_Flux_Density in erg-1 cm-2 s-1
alpha_LP = 2.1964812                                                              # LP_Index
beta_LP  = 0.07365552                                                             # LP_beta

# We integrate the dNdE curve, in units of erg-1 cm-2 s-1. When integrating, the result is the area, so 
# the unit of the result is the unit of X (erg) * the units of Y (erg-1 cm-2 s-1):
integral_flux = integrate.quad(lambda x: K_LP * ((x/E0_LP)**(-alpha_LP - beta_LP * np.log(x/E0_LP))), 
                               (0.1*u.GeV).to('erg').value, 
                               (100*u.GeV).to('erg').value)

# Result seems to make sense in terms of scale, if I compare it with the photon flux column seen in the LAT
# lightcurve repository
print("Average Photon flux: {} +/- {}".format(integral_flux[0], integral_flux[1]))
```

### Integrated Flux

```python
h = 6.626196*10**(-27) * u.erg * u.s
delta_nu = (100 -0.1)*2.418*10**(23) * u.s**(-1)    # 1 Gev = 2.418*10**(23) Hz


x_PL = E_PL                                                          # in erg

# dN/dE dE
y_PL = dNdE(x_PL, E0_PL, K_PL, alpha_PL)    # in erg-1 cm-2 s-1
print(y_PL.unit)


yPL_int = integrate.simpson(y_PL, x_PL)
print('integrated:', yPL_int)



# Doing it manually to compare the results


## Limits
inf_PL = (0.1 *u.GeV).to('erg')
sup_PL = (100 *u.GeV).to('erg')


## Integrate using the 1/3 Simpson's Rule

# x in 'erg'
def func2(x):
    y_manual = dNdE(x, E0_PL, K_PL, alpha_PL)
    return y_manual


step_PL = (inf_PL+sup_PL)/10000    # slicing in pieces
somatory_PL = 0                    # the sum of all terms
i_PL = inf_PL                      # to count

# int (dN/dE) dE
while (i_PL < sup_PL):
    a_PL = i_PL
    b_PL = i_PL+step_PL
    h_PL = (b_PL-a_PL)/2
    
    # integrating manually
    somatory_PL += (func2(a_PL)+ 4 * func2((a_PL+b_PL)/2)+ func2(b_PL)) * h_PL / 3
    i_PL += step_PL


print('sum of all terms:', somatory_PL)

print(somatory_PL)

yPL_manual = (somatory_PL.value)

print('integrated manually:', yPL_manual)
```

### Time in UTC

```python
# Functions


# time in LAT is done using MET (seconds) from January 1st, 2001 until now
# https://fermi.gsfc.nasa.gov/ssc/data/analysis/documentation/Cicerone/Cicerone_Data/Time_in_ScienceTools.html

# convert time from MET to UTC
def convert_MET_UTC(time_MET):
    
    # changing data to Unix (referencial of 1970)
    time_Unix = Time(time_MET, format='unix', scale='utc')

    # finding how many seconds are there from 1970 to January 1st, 2001
    time_difference = Time('2001-01-01', format='iso', scale='utc')
    time_difference.format = 'unix'
    time_difference.value

    # increased the diffence of time between 2001 and 1970 to my data
    time_MET_copy = np.copy(time_MET)
    time_MET_copy += time_difference.value

    # transform into year: format='iso'
    time_Unix = Time(time_MET_copy, format='unix', scale='utc')
    time_Unix.format = 'iso'
    time_Unix

    # list with only date values, year, month and day
    time_UTC = []
    for i in range(len(time_Unix.value)):
        time_UTC.append(datetime.strptime(time_Unix.value[i][:10], '%Y-%m-%d'))

    return time_UTC
```

```python
# conversion of time from MET to UTC

time2_UTC = convert_MET_UTC(time2)
time2_error_UTC = convert_MET_UTC(time2_error)
time2_upper_lim_UTC = convert_MET_UTC(time2_upper_lim)
```
