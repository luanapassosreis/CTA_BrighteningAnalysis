```python
## functions


# PARAMETERS MUST BE GIVEN IN CGS UNITS!


# returns 'erg cm-2 s-1'
def diff_flux(E, E_0, K, alpha, beta=None):

    if (beta == None):
        Y = E**2 * K * ((E/E_0)**(-alpha))
    else:
        Y = E**2 * K * ((E/E_0)**(-alpha - beta * np.log(E/E_0)))
        
    return Y

# dN/dE in 'erg-1 cm-2 s-1'
def dNdE(E, E_0, K, alpha, beta=None):
    
    if (beta == None):
        Y = K * ((E/E_0)**(-alpha))
    else:
        Y = K * ((E/E_0)**(-alpha - beta * np.log(E/E_0)))
        
    return Y
```

## 4FGL+J1256.1-0547 - LogParabola Spectrum

https://fermi.gsfc.nasa.gov/ssc/data/access/lat/LightCurveRepository/source.html?source_name=4FGL_J1256.1-0547

```python
# opening JSON file

file1 = open('4FGL+J1256.1-0547.json')

# returns JSON object as a dictionary
data1 = json.load(file1)

print(data1.keys())

# data['flux'] to select only the flux
```

```python
# LogParabola Spectrum


# 4FGL+J1256.1-0547 - LogParabola Spectrum


# PARAMETERS MUST BE GIVEN AS IT FOLLOWS:
# E: erg, E_0: erg, K: erg-1 cm-2 s-1, alpha: adimensional, ?beta: adimensional

E1     = ((np.logspace(np.log10(0.1), np.log10(100), 100) * u.GeV).to('erg')).value     # log scale of Energy in erg
E1_0   = ((466.0457 * u.MeV).to('erg')).value                                            # Pivot_Energy in erg
K1     = ((3.298064e-10 * u.MeV**-1 * u.cm**-2 * u.s**-1).to('erg-1 cm-2 s-1')).value    # LP_Flux_Density in erg-1 cm-2 s-1
alpha1 = 2.1964812                                                              # LP_Index
beta1  = 0.07365552                                                             # LP_beta


# function returns differential flux in 'erg cm-2 s-1'
Y1 = diff_flux(E1, E1_0, K1, alpha1, beta1)

plt.figure(figsize=(7,5), dpi=100)

plt.plot(((E1* u.erg).to('GeV')).value, Y1*10**(11), '+', markersize=2, color='black')
plt.plot(((E1* u.erg).to('GeV')).value, Y1*10**(11), '--', linewidth=0.4, color='black')

plt.xscale('log')
plt.yscale('log')

plt.gca().xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%i'))

plt.xlim(0.099, 120)
plt.ylim(0.43, 30)

plt.title('4FGL+J1256.1-0547 - LogParabola Spectrum', fontsize='large')
plt.ylabel(r'$\nu\ F_{\nu}$ [ $ 10^{-11} $ erg $cm^{-2}$ $s^{-1}$]', fontsize=12)
plt.xlabel('Energy (GeV)', fontsize=12)
```

```python
x1 = E1                                   # in erg
y1 = dNdE(x1, E1_0, K1, alpha1, beta1)    # in erg-1 cm-2 s-1

print('y1.unit', y1.unit)

## Limits
inf1 = (0.1 *u.GeV).to('erg')
sup1 = (100 *u.GeV).to('erg')

y1_number = integrate.simpson(y1, x1)

print('number', y1_number)





## Integrate using the 1/3 Simpson's Rule

# x in 'erg'
def func1(x):
    y_hand = dNdE(x, E1_0, K1, alpha1, beta1)
    return y_hand


step1 = (inf1+sup1)/10000    # Slicing pieces
somatory1 = 0
i1 = inf1

## Integrate dN/dE dE
while (i1 < sup1):
    a1 = i1
    b1 = i1+step1
    h1 = (b1-a1)/2
    
    #integrate
    somatory1 += (func1(a1)+ 4 * func1((a1+b1)/2)+ func1(b1)) * h1 / 3
    i1 += step1


print(somatory1)

number1 = (somatory1).value

print(number1)
```

```python
## ---------- convert time from MET to UTC ----------
def convert_MET_UTC(time_MET):
    ########## Description ##########
    # more info on time in LAT:
    # https://fermi.gsfc.nasa.gov/ssc/data/analysis/documentation/Cicerone/Cicerone_Data/Time_in_ScienceTools.html
    # ---------- Input ----------
    # time_MET : time using MET (seconds) from January 1st, 2001 until now  (1D-array)
    # ---------- Output ----------
    # time in UTC (1D-array)
    
    # changing data to Unix (referencial of 1970)
    time_Unix = Time(time_MET, format='unix', scale='utc')

    # finding how many seconds are there from 1970 to January 1st, 2001
    time_difference = Time('2001-01-01', format='iso', scale='utc')
    time_difference.format = 'unix'
    time_difference.value

    # increased the diffence of time between 2001 and 1970 to my data
    time_MET_copy = np.copy(time_MET)
    time_MET_copy += time_difference.value

    # transform into year: format='iso'
    time_Unix = Time(time_MET_copy, format='unix', scale='utc')
    time_Unix.format = 'iso'
    time_Unix

    # list with only date values, year, month and day
    time_UTC = []
    for i in range(len(time_Unix.value)):
        time_UTC.append(datetime.strptime(time_Unix.value[i][:10], '%Y-%m-%d'))

    return time_UTC
```

```python
# PowerLaw Spectrum


# 4FGL+J2256.0-2740 - PowerLaw Spectrum


# PARAMETERS MUST BE GIVEN AS IT FOLLOWS:
# E: GeV, E_0: GeV, K: GeV-1 cm-2 s-1, alpha: adimensional, ?beta: adimensional

E2     = (np.logspace(np.log10(0.1), np.log10(100), 100) * u.GeV).to('erg')      # log scale of Energy in GeV
E2_0   = (238.38518 * u.MeV).to('erg')                                           # Pivot_Energy in MeV
K2     = (5.0604333e-11 * u.MeV**-1 * u.cm**-2 * u.s**-1).to('erg-1 cm-2 s-1')   # PL_Flux_Density in Mev-1 cm-2 s-1
alpha2 = 3.0227165                                                               # PL_Index


# function returns differential flux in 'erg cm-2 s-1'
Y2 = diff_flux(E2, E2_0, K2, alpha2)


plt.figure(figsize=(7,5), dpi=100)

plt.plot(E2.to('GeV'), Y2*10**(12), '+', markersize=2, color='black')
plt.plot(E2.to('GeV'), Y2*10**(12), '--', linewidth=0.4, color='black')

plt.xscale('log')
plt.yscale('log')

plt.gca().xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))

# plt.xlim(0.099, 110)
plt.ylim(0.06, 20)

plt.title('4FGL+J2256.0-2740 - PowerLaw Spectrum', fontsize='large')
plt.ylabel(r'$\nu\ F_{\nu}$ [ $ 10^{-12} $ erg $cm^{-2}$ $s^{-1}$]', fontsize=12)
plt.xlabel('Energy (GeV)', fontsize=12)
```

```python
x2 = E2                            # in erg
y2 = dNdE(x2, E2_0, K2, alpha2)    # in erg-1 cm-2 s-1

print('y2.unit', y2.unit)

## Limits
inf2 = (0.1 *u.GeV).to('erg')
sup2 = (100 *u.GeV).to('erg')

y2_number = integrate.simpson(y2, x2)

print('number', y2_number)





## Integrate using the 1/3 Simpson's Rule

# x in 'erg'
def func2(x):
    y_hand = dNdE(x, E2_0, K2, alpha2)
    return y_hand


step2 = (inf2+sup2)/10000    # Slicing pieces
somatory2 = 0
i2 = inf2

## Integrate dN/dE dE
while (i2 < sup2):
    a2 = i2
    b2 = i2+step2
    h2 = (b2-a2)/2
    
    #integrate
    somatory2 += (func2(a2)+ 4 * func2((a2+b2)/2)+ func2(b2)) * h2 / 3
    i2 += step2


print(somatory2)

number2 = (somatory2).value

print(number2)
```
